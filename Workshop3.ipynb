```python
import pandas as pd

df = pd.read_csv("titanic.csv")
df.info()

# PART 1: Highest Fare Paid Relative to Age

# Fill missing Age values with the median age
df["Age"] = df["Age"].fillna(df["Age"].median())

# 1. Create fare_per_year = Fare / Age
df["fare_per_year"] = df["Fare"] / df["Age"]

# 2. Subset rows where fare_per_year > 5
high_fare_age = df[df["fare_per_year"] > 5]

# 3. Sort by fare_per_year (descending)
high_fare_age_srt = high_fare_age.sort_values(by="fare_per_year", ascending=False)

# 4. Select Name + fare_per_year columns
result = high_fare_age_srt[["Name", "fare_per_year"]]

# 5. View result
print(result) #Allison, Master Hudson Trevor recorded the highest fare when adjusted for age, paying roughly 164.73 per year of age.

# PART 2: Adult Male Who Paid Highest Fare Relative to Class

# 1. Create fare_per_class = Fare / Pclass
df["fare_per_class"] = df["Fare"] / df["Pclass"]

# 2. Filter adult males (Age â‰¥ 18 and Sex = male)
adult_males = df[(df["Sex"] == "male") & (df["Age"] >= 18)]

# 3. Sort by fare_per_class (descending)
adult_males_srt = adult_males.sort_values(by="fare_per_class", ascending=False)

# 4. Select Name, Age, fare_per_class
result = adult_males_srt[["Name", "Age", "fare_per_class"]]

# 5. View result
print(result) #Lesurer, Mr. Gustave J., and Cardeza, Mr. Thomas Drake Martinez paid the highest fares in their respective classes, with an estimated class-adjusted fare of about 512.33.
```

```python
import pandas as pd

df = pd.read_csv("titanic.csv")
df.info()

# PART 1: Percent of Total Fare Revenue by Passenger Class

# 1. Total fare paid by all passengers
total_fare = df["Fare"].sum()

# 2. Total fare from First Class (Pclass = 1)
total_fare_first = df[df["Pclass"] == 1]["Fare"].sum()

# 3. Total fare from Second Class (Pclass = 2)
total_fare_second = df[df["Pclass"] == 2]["Fare"].sum()

# 4. Total fare from Third Class (Pclass = 3)
total_fare_third = df[df["Pclass"] == 3]["Fare"].sum()

# Combine totals into a list
fare_totals_by_class = [total_fare_first, total_fare_second, total_fare_third]

# 5. Proportion of revenue from each class
fare_proportions = [
    total_fare_first / total_fare,
    total_fare_second / total_fare,
    total_fare_third / total_fare
]

print("Fare revenue proportions (1st, 2nd, 3rd class):")
print(fare_proportions) # Contribution of first class passenger is about 63.3% of total fare
                        # Contribution of first class passenger is about 13.2% of total fare
                        # Contribution of first class passenger is about 23.4% of total fare

# PART 2: Passenger Percent by Age Group (Child, Adult, Senior)

# Fill missing age with median
df["Age"] = df["Age"].fillna(df["Age"].median())

# 1. Create age_group column
df.loc[df["Age"] < 18, "age_group"] = "child"
df.loc[(df["Age"] >= 18) & (df["Age"] < 65), "age_group"] = "adult"
df.loc[df["Age"] >= 65, "age_group"] = "senior"

# 2. Total number of passengers
total_passengers = len(df)

# 3. Count passengers in each age group
age_group_counts = df["age_group"].value_counts()

# 4. Proportion of each age group
age_group_proportions = age_group_counts / total_passengers

# 5. Convert to percentages
age_group_percent = age_group_proportions * 100

print("\nPercentage of passengers in each age group:")
print(age_group_percent) #adults were 86.10%, children passenger were 12.69 and senor were 1.23%
```
