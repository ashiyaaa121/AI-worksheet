```python
#Problem 1
import numpy as np

# 1. Initialize an empty array with size 2x2
empty_arr = np.empty((2, 2))
print("1. Empty array:\n", empty_arr)

# 2. Initialize an all-one array with size 4x2
ones_arr = np.ones((4, 2) ,dtype=int)
print("\n2. All-one array:\n", ones_arr)

# 3. Array filled with a specific value (shape & type)
filled_arr = np.full((3, 3), fill_value=7, dtype=int)
print("\n3. Full array (3x3 filled with 7):\n", filled_arr)

# 4. Zeros with same shape and type as another array
sample_arr = np.array([[1, 2, 3], [4, 5, 6]], dtype=int)
zeros_like_arr = np.zeros_like(sample_arr)
print("\n4. Zeros like existing array:\n", zeros_like_arr)

# 5. Ones with same shape and type as another array
ones_like_arr = np.ones_like(sample_arr)
print("\n5. Ones like existing array:\n", ones_like_arr)

# 6. Convert list to numpy array
new_list = [1, 2, 3, 4]
numpy_arr = np.array(new_list)
print("\n6. List converted to numpy array:\n", numpy_arr)
```

```python
#Problem 2
import numpy as np

# 1. Create an array with values ranging from 10 to 49
array_10_49 = np.arange(10, 50)
print("1.\n", array_10_49)

# 2. Create a 3x3 matrix with values ranging from 0 to 8
matrix_3x3 = np.arange(9).reshape(3, 3)
print("\n2.\n", matrix_3x3)

# 3. Create a 3x3 identity matrix
identity_matrix = np.eye(3)
print("\n3.\n", identity_matrix)

# 4. Create a random array of size 30 and find the mean
random_array = np.random.random(30)
mean_value = random_array.mean()
print("\n4.\n", random_array)
print("Mean of the array:", mean_value)

# 5. Create a 10x10 array with random values and find min and max
random_10x10 = np.random.random((10, 10))
min_value = random_10x10.min()
max_value = random_10x10.max()
print("\n5.\n", random_10x10)
print("Min:", min_value, "Max:", max_value)

# 6. Create a zero array of size 10 and replace 5th element with 1
zero_array = np.zeros(10)
zero_array[4] = 1
print("\n6.\n", zero_array)

# 7. Reverse an array arr = [1,2,0,0,4,0]
arr = np.array([1, 2, 0, 0, 4, 0])
reversed_arr = arr[::-1]
print("\n7.\n", reversed_arr)

# 8. Create a 2D array with 1 on border and 0 inside
n, m = 5, 5
border_array = np.ones((n, m))
border_array[1:-1, 1:-1] = 0
print("\n8.\n", border_array)

# 9. Create an 8x8 matrix with a checkerboard pattern
checkerboard = np.zeros((8, 8))
checkerboard[1::2, ::2] = 1
checkerboard[::2, 1::2] = 1
print("\n9. 8x8 Checkerboard pattern:\n", checkerboard)
```

```python
#Problem 3
import numpy as np

# Given arrays
x = np.array([[1,2],[3,5]])
y = np.array([[5,6],[7,8]])
v = np.array([9,10])
w = np.array([11,12])

print("1.(x + y):")
print(x + y)

print("\n2.(x - y):")
print(x - y)

print("\n3.(3 * x):")
print(3 * x)

print("\n4.(x**2):")
print(x**2)

print("\n5(a). Dot product of v.w:")
print(np.dot(v, w))

print("\n5(b). Dot product of x.v:")
print(np.dot(x, v))

print("\n5(c). Dot product of x .y:")
print(np.dot(x, y))

print("\n6(a). Concatenate x and y along rows:")
print(np.vstack((x, y)))

print("\n6(b). Concatenate v and w along columns:")
print(np.column_stack((v, w)))

print("\n7. Concatenate x and v (Shows error explanation):")
try:
    print(np.concatenate((x, v)))
except Exception as e:
    print("Error:", e)
    print("Reason: x is 2D but v is 1D. NumPy cannot concatenate arrays with different dimensions.")
```

```python
#Problem 4
# Part 1: Matrices A and B

A = np.array([[3, 4], [7, 8]])
B = np.array([[5, 3], [2, 1]])

# Prove A * A^-1 = I
A_inv = np.linalg.inv(A)
I = np.dot(A, A_inv)
print("A * A^-1 =\n", np.round(I, 2))

# Prove AB ≠ BA
AB = np.dot(A, B)
BA = np.dot(B, A)
print("\nAB =\n", AB)
print("BA =\n", BA)

# Prove (AB)ᵀ = Bᵀ * Aᵀ
LHS = AB.T
RHS = np.dot(B.T, A.T)
print("\n\n(AB)^T =\n", LHS)
print("B^T * A^T =\n", RHS)


# Part 2: Solve linear system
# 2x - 3y + z = -1
# x - y + 2z = -3
# 3x + y - z = 9

A_sys = np.array([[2, -3, 1],
                  [1, -1, 2],
                  [3, 1, -1]])
B_sys = np.array([-1, -3, 9])

# Method 1: Using inverse
X_inv = np.dot(np.linalg.inv(A_sys), B_sys)
print("\nSolution using inverse method [x, y, z] =", np.round(X_inv, 2))

# Method 2: Using np.linalg.solve (more efficient)
X_solve = np.linalg.solve(A_sys, B_sys)
print("Solution using np.linalg.solve [x, y, z] =", np.round(X_solve, 2))
```

```python
# Experiment 4.2

import time
import numpy as np

print("1. ELEMENT-WISE ADDITION")

# Python list addition
list1 = list(range(1_000_000))
list2 = list(range(1_000_000))

start = time.time()
list_add = [list1[i] + list2[i] for i in range(1_000_000)]
end = time.time()
print("Python List Addition Time:", end - start, "seconds")

# NumPy addition
arr1 = np.arange(1_000_000)
arr2 = np.arange(1_000_000)

start = time.time()
arr_add = arr1 + arr2
end = time.time()
print("NumPy Array Addition Time:", end - start, "seconds")


print("\n2. ELEMENT-WISE MULTIPLICATION")

# Python list multiplication
start = time.time()
list_mul = [list1[i] * list2[i] for i in range(1_000_000)]
end = time.time()
print("Python List Multiplication Time:", end - start, "seconds")

# NumPy multiplication
start = time.time()
arr_mul = arr1 * arr2
end = time.time()
print("NumPy Array Multiplication Time:", end - start, "seconds")


print("\n3. DOT PRODUCT")

# Python list dot product
start = time.time()
dot_py = sum(list1[i] * list2[i] for i in range(1_000_000))
end = time.time()
print("Python List Dot Product Time:", end - start, "seconds")

# NumPy dot product
start = time.time()
dot_np = np.dot(arr1, arr2)
end = time.time()
print("NumPy Dot Product Time:", end - start, "seconds")

print("\n4. MATRIX MULTIPLICATION (1000 x 1000)")

# Python list matrix multiplication
matrix1 = [[1]*1000 for _ in range(1000)]
matrix2 = [[1]*1000 for _ in range(1000)]

start = time.time()
result_matrix = [
    [sum(matrix1[i][k] * matrix2[k][j] for k in range(1000))
    for j in range(1000)]
    for i in range(1000)
]
end = time.time()
print("Python List Matrix Multiplication Time:", end - start, "seconds")

# NumPy matrix multiplication
np_matrix1 = np.ones((1000, 1000))
np_matrix2 = np.ones((1000, 1000))

start = time.time()
np_result = np.dot(np_matrix1, np_matrix2)
end = time.time()
print("NumPy Matrix Multiplication Time:", end - start, "seconds")
```
